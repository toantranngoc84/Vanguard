# 6 Merge dexp and web / Create columns time difference and time cumulativ

## Merge dexp_df and web_df

```python
merged_df = pd.merge(dexp_df, web_df, on='client_id', how='left')
display(merged_df.head())
```

```python
display(merged_df.shape)
```

## Create Column time difference

```python
merged_df['date_time'] = pd.to_datetime(merged_df['date_time'])

# Sort by visit_id and date_time to ensure the correct order for time difference calculation
merged_df = merged_df.sort_values(by=['visit_id', 'date_time'])

# Initialize the time_diff column
merged_df['time_diff'] = 0

# Group by visit_id to calculate time differences within each visit_id
for visit_id, group in merged_df.groupby('visit_id'):
    previous_time = None
    for idx, row in group.iterrows():
        if row['process_step'] == 'start' and previous_time is None:
            merged_df.at[idx, 'time_diff'] = 0
        elif previous_time is not None:
            merged_df.at[idx, 'time_diff'] = (row['date_time'] - previous_time).total_seconds()
        previous_time = row['date_time']
```

## Create Column time cumulativ

```python
# Create the cumulative sum column
merged_df['time_cum'] = merged_df.groupby(['client_id', 'visit_id'])['time_diff'].cumsum()
```

## Format time to seconds

```python
def seconds_to_min_sec(seconds):
    minutes = seconds // 60
    sec = seconds % 60
    return f"{minutes}:{sec:02d}"

# Apply the function to the time_cum column
merged_df['time_cum'] = merged_df['time_cum'].apply(seconds_to_min_sec)
merged_df['time_diff'] = merged_df['time_diff'].apply(seconds_to_min_sec)
```

# 8 Create DF with process start and end

```python
# sort by client_id,visit_id, date and time
merged_sorted_df = merged_df.sort_values(by=['client_id', 'visit_id', 'date', 'time'], ascending = [True, True, True, True])
# define colum process_step_stop by shifting column process_step by one row
merged_sorted_df['process_step_stop'] = merged_sorted_df.groupby('visitor_id')['process_step'].shift(-1, fill_value='END')
# define colum date_time_end by shifting column date_time by one row
merged_sorted_df['date_time_end'] = merged_sorted_df.groupby('visitor_id')['date_time'].shift(-1, fill_value='END')
# shift column time_diff by one row
merged_sorted_df['time_diff'] = merged_sorted_df.groupby('visitor_id')['time_diff'].shift(-1, fill_value='END')
# shift column time_cum by one row
merged_sorted_df['time_cum'] = merged_sorted_df.groupby('visitor_id')['time_cum'].shift(-1, fill_value='END')
#rename process_step to process_step_start
#rename date_time to date_time_start
merged_sorted_df = merged_sorted_df.rename(columns = {'process_step':'process_step_start','date_time':'date_time_start'})

print(merged_sorted_df.shape)
display(merged_sorted_df.head())
```

```python
filtered_merge_df = merged_sorted_df

# filter all rows with values END (especially for process_step)
filtered_merge_df = filtered_merge_df[~filtered_merge_df.apply(lambda row: row.astype(str).str.contains('END').any(), axis=1)]
filtered_merge_df.reset_index(drop=True, inplace=True)
display(filtered_merge_df.head())
print(filtered_merge_df.shape)
```

## Create Column Step Counts

```python
# define order of the steps
correct_order = {'start': 1, 'step_1': 2, 'step_2': 3, 'step_3': 4, 'confirm': 5}

# Map the process steps to their respective numerical values
filtered_merge_df['start_order'] = filtered_merge_df['process_step_start'].map(correct_order)
filtered_merge_df['stop_order'] = filtered_merge_df['process_step_stop'].map(correct_order)

# Calculate step_count based on the comparison
def calculate_step_count(row):
    if row['start_order'] < row['stop_order']:
        return 1
    elif row['start_order'] > row['stop_order']:
        return -1
    else:
        return 0

filtered_merge_df['step_count'] = filtered_merge_df.apply(calculate_step_count, axis=1)

# Drop the temporary columns used for mapping (optional)
filtered_merge_df.drop(['start_order', 'stop_order'], axis=1, inplace=True)

# Display the updated dataframe
display(filtered_merge_df.head())
```

## Create columns for counting positive, negative and equal steps

```python
# Resetting the index of the DataFrame
filtered_merge_df = filtered_merge_df.reset_index(drop=True)

# Function to add positive_step and negative_step columns
def add_step_flags(group):
    group['positive_step'] = (group['step_count'] == 1).astype(int)
    group['negative_step'] = (group['step_count'] == -1).astype(int)
    group['equal_step'] = (group['step_count'] == 0).astype(int)
    return group

# Apply the function to each group
filtered_merge_df = filtered_merge_df.groupby(['client_id', 'visit_id'], as_index=False, group_keys=False).apply(add_step_flags)

display(filtered_merge_df.head())
```

```python
# adding cumulative sum columns for ste_cum. pos_step_cum,neg_step_cum, equ_step_cum
filtered_merge_df['step_cum'] = filtered_merge_df.groupby(['client_id', 'visit_id'])['step_count'].cumsum()
filtered_merge_df['pos_step_cum'] = filtered_merge_df.groupby(['client_id', 'visit_id'])['positive_step'].cumsum()
filtered_merge_df['neg_step_cum'] = filtered_merge_df.groupby(['client_id', 'visit_id'])['negative_step'].cumsum()
filtered_merge_df['equ_step_cum'] = filtered_merge_df.groupby(['client_id', 'visit_id'])['equal_step'].cumsum()

# filter out all rows with 'process_step_start' != 'confirm'
filtered_merge_df= filtered_merge_df[filtered_merge_df['process_step_start'] != 'confirm']
```

```python
completion_data = {
    'Group': ['Control (UI_old)', 'Test (UI_new)'],
    'Completed': [17095, 23894],
    'Total': [26194, 34293]
}

completion_df = pd.DataFrame(completion_data)

# Calculate completion rates
completion_df['Completion Rate'] = completion_df['Completed'] / completion_df['Total']
completion_df['Completion Rate (%)'] = (completion_df['Completion Rate'] * 100).round(0)

# Calculate the difference in completion rates
completion_rate_difference = (completion_df['Completion Rate (%)'].iloc[1] - completion_df['Completion Rate (%)'].iloc[0]).round(0)

# Plot with the difference highlighted
plt.figure(figsize=(10, 6))
completion_barplot = sns.barplot(x='Group', y='Completion Rate (%)', data=completion_df, hue='Group', dodge=False, palette='viridis', legend=False)
plt.ylim(0, max(completion_df['Completion Rate (%)']) + 5)
plt.title('Completion Rates for Control and Test Groups')
plt.xlabel('Group')
plt.ylabel('Completion Rate (%)')

# Annotate the bars with completion rates
for index, row in completion_df.iterrows():
    completion_barplot.text(index, row['Completion Rate (%)'] + 1, f"{row['Completion Rate (%)']:.0f}%", color='black', ha="center")

# Highlight the difference
x = [0, 1]
y = completion_df['Completion Rate (%)']
plt.plot(x, y, linestyle='--', color='gray')
plt.fill_between(x, y[0], y[1], color='orange', alpha=0.3)

# Annotate the difference percentage
completion_barplot.annotate(f'Increase: {completion_rate_difference:.0f}%', 
                            xy=(0.5, (y[0] + y[1]) / 2), 
                            xytext=(0.5, (y[0] + y[1]) / 2 + 2),
                            arrowprops=dict(arrowstyle="->", color='black'),
                            ha='center', color='black', fontsize=12)

plt.show()
```
